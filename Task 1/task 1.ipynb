```python
# ==============================================================================
# Task 1: Iris Data Exploration & Visualization
# ==============================================================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df_iris = pd.read_csv("iris.csv")

# 1. Structure Check
print("=== Task 1: Structure ===")
print(f"Shape: {df_iris.shape}")
print(f"Columns: {df_iris.columns.tolist()}")
print(df_iris.head(), "\n")

# 2. Visualizations
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Task 1: Iris Dataset Exploratory Analysis", fontsize=14, weight='bold')

# Scatter Plot
sns.scatterplot(data=df_iris, x='petal_length', y='petal_width', hue='species', ax=axes[0], palette='deep')
axes[0].set_title("Petal Length vs Width")

# Histogram
sns.histplot(data=df_iris, x='sepal_length', hue='species', kde=True, element='step', ax=axes[1], palette='muted')
axes[1].set_title("Sepal Length Distribution")

# Box Plot
sns.boxplot(data=df_iris, x='species', y='sepal_width', ax=axes[2], palette='Pastel1', hue='species', legend=False)
axes[2].set_title("Sepal Width Spread & Outliers")

plt.tight_layout()
plt.show()

print("Task 1 Insights: Iris-setosa is completely linearly separable using petal metrics.")
```
